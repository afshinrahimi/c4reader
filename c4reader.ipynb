```python
!pip install datasets
!pip install transformers
!pip install spacy
!pip install nltk
import spacy
import json
from datasets import load_dataset
import spacy
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
dataset = load_dataset('c4', 'en', streaming=True)
i = 0
smell_words = set(['smell', 'smells', 'smelling', 'smelled', 'stink', 'scent', 'aroma'])
#smell_words = set(['smells', 'like'])
for instance in dataset['train']:
      text = instance['text']
      sentences = sent_tokenize(text)
      for j, sentence in enumerate(sentences):
          tokens = set(word_tokenize(sentence))
          if len(tokens & smell_words) > 1: # if there is any smell word in the text
              prev_sentence = sentences[j-1] if j > 0 else ""
              print(i, prev_sentence, sentence)
              #print(i, sentence)
              i += 1
              if i > 10:
                  break
```
